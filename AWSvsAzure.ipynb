```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aws = pd.read_csv('recorded_performance_aws2.csv', names = ['word', 'confScore', 'type', 'resolutionLevel', 'service'])
azure = pd.read_csv('recorded_performance_azure2.csv', names = ['word', 'confScoreAzure', 'typeAzure', 'resolutionLevel', 'serviceAzure'])
```

```python
azure = azure[['word', 'confScoreAzure', 'resolutionLevel', 'serviceAzure']]
azure['confScoreAzure'] = azure['confScoreAzure']*100
aws = aws.round({'confScore':1})
```

```python
print('Shape of AWS result', aws.shape)
aws.head(45)
```

```python
print('Shape of Azure result', azure.shape)
azure.head(45)
```

```python
azure3 = azure.loc[azure['resolutionLevel'] == 3]
azure2 = azure.loc[azure['resolutionLevel'] == 2]
azure1 = azure.loc[azure['resolutionLevel'] == 1]
```

```python
aws3 = aws.loc[aws['resolutionLevel'] == 3]
aws2 = aws.loc[aws['resolutionLevel'] == 2]
aws1 = aws.loc[aws['resolutionLevel'] == 1]
```

```python
print('Number of words recognised using lowest resolution AWS:', aws1.shape[0])
print('Number of words recognised using medium resolution AWS:', aws2.shape[0])
print('Number of words recognised using original AWS:', aws3.shape[0])
print('===========================')
print('Number of words recognised using lowest resolution Azure:', azure1.shape[0])
print('Number of words recognised using medium resolution Azure:', azure2.shape[0])
print('Number of words recognised using original Azure:', azure3.shape[0])
```

```python
azure_3 = azure3.groupby(['word'], as_index = False).mean()
aws_3 = aws3.groupby(['word'], as_index = False).mean()
df3 = aws_3.merge(azure_3, on = ['word'], how = 'inner')
print('Shape 3', df3.shape)

azure_2 = azure2.groupby(['word'], as_index = False).mean()
aws_2 = aws2.groupby(['word'], as_index = False).mean()
df2 = aws_2.merge(azure_2, on = ['word'], how = 'inner')
print('Shape 2', df2.shape)

azure_1 = azure1.groupby(['word'], as_index = False).mean()
aws_1 = aws1.groupby(['word'], as_index = False).mean()
df1 = aws_1.merge(azure_1, on = ['word'], how = 'inner')
print('Shape 1', df1.shape)
```

```python
# line 1 points
x1 = range(165)
y1 = df3.confScore
# plotting the line 1 points 
plt.scatter(x1, y1, label = "AWS", alpha = 0.5)
# line 2 points
x2 = x1
y2 = df3.confScoreAzure
# plotting the line 2 points 
plt.scatter(x2, y2, label = "Azure", alpha = 0.5)
plt.xlabel('words')
# Set the y axis label of the current axis.
plt.ylabel('confidence scores')
# Set a title of the current axes.
plt.title('Original Scan')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.savefig('Original')
plt.show()
```
